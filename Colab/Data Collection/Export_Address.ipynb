```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/"

contract_csv = "Dataset/Contract_Address.csv"
```

```python
arbitrum_sequencer = "0xa4b10ac61e79ea1e150df70b8dda53391928fd14"
ethermine = "0xea674fdde714fd979de3edf0f56aa9716b898ec8"
```

```python
values_to_exclude = [ethermine, arbitrum_sequencer]
```

```python
df_contract = pd.read_csv(path + contract_csv)
```

```python
def read_chunk_csv(file_name):
    try:
        df = pd.read_csv(path + file_name, usecols=["from", "to"])

        df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

        series_sample = pd.concat([df["from"], df["to"]], ignore_index=True).drop_duplicates().dropna()
        return pd.DataFrame(series_sample, columns=["Address"]).reset_index(drop=True)

    except FileNotFoundError:
        print(f'Error Not Found File {file_name}')
```

```python
def export_address(dataframe, count):
  dataframe.to_csv(path + export_file_name + f'{count}.csv', index=False)
```

```python
def processing_export(file_name, count):
    print(f'{file_name}')
    try:
        df_sample = read_chunk_csv(file_name).drop_duplicates()

        mask = df_sample['Address'].isin(df_contract['Address'])
        df_sample = df_sample[~mask]

        mask_address = df_sample['Address'].isin(values_to_exclude)
        df_sample = df_sample[~mask_address]

        export_address(df_sample, count)
        print('Finish!')
    except FileNotFoundError:
        print(f'Error Not Found File {file_name}')
    except Exception as e:
        print(f'Error {e}')
```

```python
csv_file_name = 'Data/15000000to15249999/15000000to15249999_BlockTransaction_chunk_'

export_file_name = "Dataset/16750000to16999999/16750000to16999999_Address_"
```

```python
for i in range(19):
  processing_export(csv_file_name + str(i) + ".csv", i)
```
