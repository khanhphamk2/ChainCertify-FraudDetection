```python
import zipfile
import pandas as pd
import io
```

```python
# columns_to_read = ["from", "to"]

path = "/content/drive/MyDrive/Colab Notebooks/Data/"

blocks = ["16250000to16499999"]

file_detail = '_BlockTransaction'
```

```python
def process_chunks_from_zip(path_new, file_name, size=2000000):
    try:
      with zipfile.ZipFile(path_new + file_name + ".zip", 'r') as zip_ref:
        print(f'File {file_name}.zip is processing...')
        chunks = pd.read_csv(zip_ref.open((file_name + ".csv")), chunksize=size, index_col=False, low_memory=False)
        i = 0
        for chunk in chunks:
            print(f'Chunk {i} processing...')

            # df_chunk = pd.DataFrame(chunk)

            chunk.to_csv(path_new + f'{file_name}_chunk_{i}.csv', index=False)

            print(f'Chunk {i} Finish!')

            i += 1
    except Exception as e:
      print(f'Error: {e}')
```

```python
paths_and_names = [(path + b + "/", b + file_detail) for b in blocks]
```

```python
for path_new, f_name in paths_and_names:
    process_chunks_from_zip(path_new, f_name)
```
