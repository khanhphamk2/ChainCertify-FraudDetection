### Functions

```python
def count_csv_files(directory):
    folder_path =  path + directory

    csv_files = glob.glob(folder_path + '/*.csv')

    csv_count = len(csv_files)

    return csv_count
```

```python
def export_address(dataframe, range_block):
  dataframe.to_csv(export_path + f'{range_block}_address.csv', index=False)
```

```python
def read_chunk_csv(file_name):
    try:
        df = pd.read_csv(path + file_name, usecols=["from", "to"])

        df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

        series_sample = pd.concat([df["from"], df["to"]], ignore_index=True).drop_duplicates().dropna()
        return pd.DataFrame(series_sample, columns=["Address"]).reset_index(drop=True)

    except FileNotFoundError:
        print(f'Error Not Found File {file_name}')
```

```python
def read_csv(filename):
    try:
        df = pd.read_csv(path + filename)
        df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
        return df
    except Exception as e:
        print(f'Error: {e}')
```

```python
def get_address_from_chunk(range_block, count):
    try:
        df_sample = read_csv(f'{range_block}/{range_block}_Address_{count}.csv').drop_duplicates()

        mask = df_sample['Address'].isin(df_contract['Address'])
        df_sample = df_sample[~mask]

        mask_address = df_sample['Address'].isin(values_to_exclude)
        df_sample = df_sample[~mask_address]

        return df_sample
    except Exception as e:
        print(f'Error {e}')
```

```python
def get_addresses(range_block, df_sample, num_of_file):
    try:
        for i in range(num_of_file):
            print(f'Processing Chunk {i}...')

            df_draft = get_address_from_chunk(range_block, i)

            df_sample = pd.concat([df_sample, df_draft], ignore_index=True).drop_duplicates()

            print(f'Chunk {i} Successful!')

        return df_sample
    except Exception as e:
        print(f'Error {e}')
```

```python
def processing_export():
    try:
        for range_block in blocks:
            num_of_file = count_csv_files(range_block)

            print(f'{range_block} - {num_of_file }')

            df = pd.DataFrame(columns=['Address'])

            df = get_addresses(range_block, df, num_of_file)

            export_address(df, range_block)

            print(f'{range_block} Finish!')
    except FileNotFoundError:
        print(f'Error Not Found File {range_block}')
    except Exception as e:
        print(f'Error {e}')
```

### Execution

```python
processing_export()
```
